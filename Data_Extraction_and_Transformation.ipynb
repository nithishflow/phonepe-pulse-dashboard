```python
import pandas as pd
import json
import os



#aggregated insurance

path1="D:/project 1/Data/data/aggregated/insurance/country/india/state/"
Agg_insu_list=os.listdir(path1)
Agg_insu_list




clm1={'State':[], 'Year':[],'Quater':[],'Transacion_type':[], 'Transacion_count':[], 'Transacion_amount':[]}

for i in Agg_insu_list:
    p_i=path1+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            C=json.load(Data)
            for z in C['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              clm1['Transacion_type'].append(Name)
              clm1['Transacion_count'].append(count)
              clm1['Transacion_amount'].append(amount)
              clm1['State'].append(i)
              clm1['Year'].append(j)
              clm1['Quater'].append(int(k.strip('.json')))

Agg_insu=pd.DataFrame(clm1)


Agg_insu['State']=Agg_insu['State'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
Agg_insu['State']=Agg_insu['State'].str.replace('-',' ')
Agg_insu['State']=Agg_insu['State'].str.title()
Agg_insu['State']=Agg_insu['State'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')

```
